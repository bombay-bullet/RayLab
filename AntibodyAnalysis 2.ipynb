```python
def list_2_fasta(lst,names,name):
    peptide_file = open((name + '.fasta'), 'w+')
    out = '\n'.join([">" + str(i+1) + "_" + names[i] + "\n" + j for i,j in enumerate(lst)])
    peptide_file.write(out)
    peptide_file.close()
```
